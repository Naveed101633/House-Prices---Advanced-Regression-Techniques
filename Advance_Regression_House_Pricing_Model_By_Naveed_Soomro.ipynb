### 2.1. distplot
**Higher Bar in Histogram:**

The height of each bar in the histogram represents the number of houses (frequency) that have sale prices within a specific range (bin).

A higher bar means more houses are sold within that price range.
Price Range $100,000 to $200,000:

If the highest bar lies between $100,000 and

<a id = "kdeplot"></a>
### 2.2. kdeplot
***Kernel Density Estimation (KDE)*** is a way to estimate the probability density function of a continuous random variable. It is used for ***non-parametric*** analysis. Setting the `hist` flag to False in `distplot` will yield the KDE plot. For bivariate distribution, we can plot a kde by using `jointplot()`. Pass value `‘kde’` to the parameter `kind` to plot kernel plot.

**Note:** `distplot(data)` is used to visualize the ***parametric distribution*** of data. It plot both KDE and histogram on the same figure.

```python
sns.distplot(full_data['SalePrice'], hist=False)
plt.show()
```

**Interpretation:**
**Density:**

Density values on the y-axis show how common certain basement sizes are, relative to the overall distribution.

Higher density values indicate more houses with basement sizes in that range.
TotalBsmtSF:

*The range of basement sizes is shown on the x-axis, from small to large basements.*

**Understanding the Distribution:**
**Peaks in KDE**: Where the KDE curve peaks, you have the highest density of houses, meaning these basement sizes are the most common.

**Spread of Data**: The width of the KDE curve and the spread of the histogram bars show how varied the basement sizes are. A wide spread indicates a large variety of basement sizes.

**Skewness:** If the KDE curve has a long tail on one side, it indicates skewness. For example, a long right tail suggests a few houses with very large basements.

**Practical Example**:
**High Density:** If the KDE curve peaks around 1,000 square feet, it means that most houses have basements around this size.

**Low Densit**y: If density is low beyond 2,000 square feet, it means very few houses have basements larger than this size.

**Skewness**: If the curve tails off to the right, it suggests that while most basements are relatively small, there are some houses with exceptionally large basements.

```python
plt.figure(figsize=(8,8))
sns.distplot(full_data['TotalBsmtSF'])
plt.show()
```

**Visualizing Skewness in Your Plot:**
***Right Skew (Positive Skew)***:
If the KDE curve has a longer tail to the right side, it means the distribution is positively skewed.
This indicates there are a few houses with much larger basements compared to the rest.

***Left Skew (Negative Skew):***
If the KDE curve has a longer tail to the left side, it means the distribution is negatively skewed.
This indicates there are a few houses with much smaller basements compared to the rest.

### 3.1. relplot
Figure-level interface for drawing relational plots onto a FacetGrid.

The function `relplot()` is named that way because it is designed to visualize many different statistical relationships. ### 3.3. lineplot
Draw a line plot with possibility of several semantic groupings.

The relationship between `x` and `y` can be shown for different subsets of the data using the `hue, size`, and `style` parameters. These parameters control what visual semantics are used to identify the different subsets. It is possible to show up to three dimensions independently by using all three semantic types, but this style of plot can be hard to interpret and is often ineffective. Using redundant semantics (i.e. both **hue** and **style** for the same variable) can be helpful for making graphics more accessible.

The default treatment of the hue (and to a lesser extent, size) semantic, if present, depends on whether the variable is inferred to represent **“numeric” or “categorical”** data. In particular, numeric variables are represented with a sequential colormap by default, and the legend entries show regular **“ticks”** with values that may or may not exist in the data. This behavior can be controlled through various parameters.

By default, the plot aggregates over multiple `y` values at each value of `x` and shows an estimate of the central tendency and a confidence interval for that estimate.

Now let's plot the correlation matrix of our data with a heatmap.

**Correlation Values:**

The correlation coefficient ranges from -1 to 1.
+1: Perfect positive correlation. As one variable increases, the other also increases.

-1: Perfect negative correlation. As one variable increases, the other decreases.

0: No correlation. The variables do not have a linear relationship.
Important Features:

Features with high positive correlation values (close to +1) with the target variable (SalePrice) are important because they increase as the target variable increases.

Features with high negative correlation values (close to -1) are also important because they decrease as the target variable increases.
